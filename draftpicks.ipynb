```python
import pandas
df = pandas.read_csv("1976_to_2015_Draftees_edit2.csv")
```

```python
df
```

```python
firstround = df[df['Pk'] <= 30]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(20, 8))
sns.boxplot(x="Pk", y="VORP", data=firstround, ax=ax)
```

```python
f, ax = plt.subplots(figsize=(20, 8))
sns.stripplot(x='Pk', y='VORP', data=firstround, ax=ax)
```

```python
import altair as alt
```

```python
chart = alt.Chart(firstround).mark_boxplot(extent=1).encode(
    x='Pk:O',
    y='VORP:Q',
    color='average(VORP):Q',
    tooltip='Player:N'
)
chart
```

```python
chart.save('index.html')
```
